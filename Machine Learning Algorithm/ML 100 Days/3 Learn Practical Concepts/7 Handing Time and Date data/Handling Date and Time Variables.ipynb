# Working with Dates

```python
# coverting to datetime datatype
date['date'] = pd.to_datetime(date['date'])
```

```python
date.info()
```

```python
date
```

# 1.Extract Year

```python
date['data_year'] = date['date'].dt.year
date
```

# 2.Extract Month

```python
date['data_month_no'] = date['date'].dt.month
date.head()
```

# 3. Extract Month

```python
date['data_month_name'] = date['date'].dt.month_name()
date.head()
```

# 4. Extract Days

```python
date['date_day'] = date['date'].dt.day
date.head()
```

# 5. Day of week

```python
date['date_dow'] = date['date'].dt.dayofweek
date.head()
```

# Day of week - Name

```python
date['date_dow_name'] = date['date'].dt.day_name()
date.head()
```

```python
# is weekend?

date['date_is_weekend'] = np.where(date['date_dow_name'].isin(['Sunday', 'Saturday']), 1,0)

date.drop(columns=['product_id','city_id','orders']).head()
```

# Extract week of the year

```python
date['date_week'] = date['date'].dt.week
```

```python
date['quarter'] = date['date'].dt.quarter
date.head()
```

# Extract Semester

```python
date['semester'] = np.where(date['quarter'].isin([1,2]), 1, 2)
date.head()
```

# Extract Time eleapsed between dates

```python
import datetime

today = datetime.datetime.today()
today
```

```python
today - date['date']
```

```python
(today - date['date']).dt.days
```

```python
# Months passed
np.round((today -date['date']) / np.timedelta64(1, 'M'),0)
```
