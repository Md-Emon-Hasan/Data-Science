# resize your graph

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]
# resize yuor graph
plt.figure(figsize=(5,3))
plt.plot(x,y,'r.--',label='2x')
x2 = np.arange(0,4,0.5)
# use sorted notaion
plt.plot(x2[:6],x2[:6]**2,'g',label='x^2')
plt.plot(x2[5:],x2[5:]**2,'g--')
plt.title('Our First Graph',fontdict={'fontname':'Comic Sans MS','fontsize':20})
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,7,7.5,10])
plt.legend()
plt.show()
```

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]
# resize yuor graph
plt.figure(figsize=(5,3),dpi=300)
plt.plot(x,y,'r.--',label='2x')
x2 = np.arange(0,4,0.5)
# use sorted notaion
plt.plot(x2[:6],x2[:6]**2,'g',label='x^2')
plt.plot(x2[5:],x2[5:]**2,'g--')
plt.title('Our First Graph',fontdict={'fontname':'Comic Sans MS','fontsize':20})
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,7,7.5,10])
plt.legend()
plt.show()
```

# save graph

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]
# resize yuor graph
plt.figure(figsize=(5,1.5),dpi=300)
# use sorted notaion
# fmt = '[color][marker][line]'
plt.plot(x,y,'r.--',label='2x')
# select intervel we want to plot point at
x2 = np.arange(0,4,0.5)
# plot part of the graph
plt.plot(x2[:6],x2[:6]**2,'g',label='x^2')
plt.plot(x2[5:],x2[5:]**2,'g--')
plt.title('Our First Graph',fontdict={'fontname':'Comic Sans MS','fontsize':20})
# x and y lebels
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
# x and y lebels ticmarks (specify font parameters with fontdict)
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,7,7.5,10])
# add a legend
plt.legend()
# save my graph
plt.savefig('mygraph.png',dpi=300)
# show plt
plt.show()
```

# boxplot style

```python
import matplotlib.pyplot as plt
fifa = pd.read_csv('fifa_data.csv')
plt.figure(figsize=(5,8))
barcelona = fifa.loc[fifa.Club == "FC Barcelona"]['Overall']
madrid = fifa.loc[fifa.Club == "Real Madrid"]['Overall']
revs = fifa.loc[fifa.Club == 'New England Revolution']['Overall']
lebels = ['FC Barcelona','Real Madrid','New England Revolution']
plt.title('Professional soccer Team Comparison')
plt.ylabel('FIFA Overall Rating')
boxes = plt.boxplot([barcelona, madrid, revs],labels=lebels,patch_artist=True)
for box in boxes['boxes']:
    # set edge color
    box.set(color='lime',linewidth=2)
    # change fill color
    box.set(facecolor='green')
plt.show()
```

```python
import matplotlib.pyplot as plt
fifa = pd.read_csv('fifa_data.csv')
plt.figure(figsize=(5,8))
barcelona = fifa.loc[fifa.Club == "FC Barcelona"]['Overall']
madrid = fifa.loc[fifa.Club == "Real Madrid"]['Overall']
revs = fifa.loc[fifa.Club == 'New England Revolution']['Overall']
lebels = ['FC Barcelona','Real Madrid','New England Revolution']
plt.title('Professional soccer Team Comparison')
plt.ylabel('FIFA Overall Rating')
boxes = plt.boxplot([barcelona, madrid, revs],labels=lebels,patch_artist=True,medianprops={'linewidth':2})
for box in boxes['boxes']:
    # set edge color
    box.set(color='lime',linewidth=2)
    # change fill color
    box.set(facecolor='green')
plt.show()
```

# Scatter Plot

# 3D Scatter Plot
